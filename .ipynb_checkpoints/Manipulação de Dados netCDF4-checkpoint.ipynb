```python
#Use ncattrs() to get the attributes

#Use getattr(data, attribute_name) to retrieve the attributes

gattrs = dt_45_2055.ncattrs() 

print gattrs
```

```python
# Como levantar os diferentes atributos e seus respectivos detalhes do objeto

for key in gattrs:

    print ("global attribute["+key+"]=" + str(getattr(dt_45_2055, key)))
```

```python
# Query somente dos atributos dos objetos

for key in gattrs:

    print ("global attribute["+key+"]")
```

```python
# Notar que a função + é essencial para a atribuição contínua dos atributos de gattrs à função print

for key in gattrs:
    print ("global attribute: +key+ = {0}".format(str(getattr(dt_45_2055, key))))
```

# Operando os dados do arquivo:

```python
#extração de uma variável de um objeto ncfile e convertendo a um array


lat = dt_45_2055.variables['lat'][:]
long = dt_45_2055.variables['lon'][:]
```

```python
#Atribuindo variáveis do netcdf a arrays numpy:



varlist = ['histclim', 'anomaly', 'histstddev', 'varratio']
                           
#Definindo variáveis

arr45_2055 =[]
arr85_2055=[]
arr45_2099=[]
arr85_2099=[]



for x in varlist:
    arr45_2055 = dt_45_2055.variables[x]
    arr85_2055 = dt_85_2055.variables[x]
    arr45_2099 = dt_45_2099.variables[x]
    arr85_2099 = dt_85_2099.variables[x]

    
    
#Verificando as dimensões da array:

print (arr45_2055.dimensions[:])


print (arr45_2055.dimensions[0])
```

```python
#Convertendo tudo em array:

arr45_2055 = np.asarray(arr45_2055)
arr85_2055 = np.asarray(arr85_2055)
arr45_2099 = np.asarray(arr45_2099)
arr85_2099 = np.asarray(arr85_2099)
```

```python
#Invertendo o eixo da longitude:

arr45_2055 = arr45_2055[:,::-1]
arr85_2055 = arr85_2055[:,::-1]
arr45_2099 = arr45_2099[:,::-1]
arr85_2099 = arr85_2099[:,::-1]



arr45_2055 = np.where(arr45_2055>1e10,np.nan,arr45_2055)
arr85_2055 = np.where(arr85_2055>1e10,np.nan,arr85_2055)
arr45_2099 = np.where(arr45_2099>1e10,np.nan,arr45_2099)
arr85_2099 = np.where(arr85_2099>1e10,np.nan,arr85_2099) 
```

```python
np.shape(arr85_2055)
```

```python
#Correção da direção das matrizes:
#arr45_2055 = arr45_2055[:,:]
#arr85_2055 = arr85_2055[:,:]
#arr45_2099 = arr45_2099[:,:]
#arr85_2099 = arr85_2099[:,:]


```

```python
#Extração de nulos e continentes:

arr45_2055 = np.where(arr45_2055>1e9,np.nan,arr45_2055)
arr85_2055 = np.where(arr85_2055>1e9,np.nan,arr85_2055)
arr45_2099 = np.where(arr45_2099>1e9,np.nan,arr45_2099)
arr85_2099 = np.where(arr85_2099>1e9,np.nan,arr85_2099) 
```

```python
#Atribuição de variáveis:

lons = dt_45_2055.variables['lon'][:]
lats = dt_45_2055.variables['lat'][:]
Anomaly_45_2055 = dt_45_2055.variables['anomaly'][:]
Anomaly_85_2055 = dt_85_2055.variables['anomaly'][:]
Anomaly_45_2099 = dt_45_2099.variables['anomaly'][:]
Anomaly_85_2099 = dt_85_2099.variables['anomaly'][:]

Anomaly_units = dt_45_2055.variables['anomaly'].units

print Anomaly_units

Mean_45_2055 = dt_45_2055.variables['histclim'][:]


lon_0 = lons.mean()
lat_0 = lats.mean()


# we setup a Basemap instance, specifying our desired map and projection settings

m = Basemap (resolution='l',projection='cea',\
            lat_ts=40,lat_0=lat_0,lon_0=lon_0)
```

```python
#When we give this Basemap instance our coordinate variables, it returns our plotting coordinates. This is how basemap knows where to put our projected data on the map

lon, lat = np.meshgrid(lons, lats)
xi, yi = m(lon, lat)
```

```python
#Salvando memória do computador:

dt_45_2055.close()
dt_85_2055.close()
dt_45_2099.close()
dt_85_2099.close()
```

# Now, we can plot the data using one of the available plot types (pcolor, pcolormesh, contour, contourf, scatter, etc.). Here we use pcolor. Gridlines, colorbars, and axis labels can also be added at this point.



```python
# Plot Data
cs_45_2055 = m.pcolor(xi,yi,np.squeeze(Anomaly_45_2055))


# Add Grid Lines
m.drawparallels(np.arange(-80., 81., 20.), labels=[1,0,0,0], fontsize=10)
m.drawmeridians(np.arange(-180., 180., 61.), labels=[0,0,0,1], fontsize=10)


# Add Coastlines, States, and Country Boundaries
m.drawcoastlines()
m.drawstates()
m.drawcountries()

# Add Colorbar
cbar = m.colorbar(cs_45_2055, location='bottom', pad="2%")
cbar.set_label(Anomaly_units)

# Add Title
plt.title('Anomalia do modelo 4.5 para 2055')


plt.show()
```
